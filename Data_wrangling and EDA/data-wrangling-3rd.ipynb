# 5. Univariate Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# lets create a function for a simpler design 
def plot_numerical_features(df, numerical_features):
    for feature in numerical_features:
        # Histogram
        plt.figure(figsize=(8, 5))
        sns.histplot(df[feature], kde=True, color='skyblue')
        plt.title(f'Histogram of {feature}')
        plt.xlabel(feature)
        plt.ylabel('Frequency')
        plt.show()

        # Box plot
        plt.figure(figsize=(8, 5))
        sns.boxplot(data=df, y=feature, color='lightgreen')
        plt.title(f'Box plot of {feature}')
        plt.ylabel(feature)
        plt.show()

        # KDE plot
        plt.figure(figsize=(8, 5))
        sns.kdeplot(df[feature], shade=True, color='salmon')
        plt.title(f'Kernel Density Estimate plot of {feature}')
        plt.xlabel(feature)
        plt.ylabel('Density')
        plt.show()
```

```python
numerical_features = ['PERSONCOUNT', 'PEDCOUNT', 'PEDCYLCOUNT', 'VEHCOUNT', 'INJURIES', 'SERIOUSINJURIES', 'FATALITIES']
```

```python
plot_numerical_features(df, numerical_features)   #function call to plot
```

```python
# lets create similar function for categorical
def plot_categorical_features(df, categorical_features):
    for feature in categorical_features:
        # Bar plot
        plt.figure(figsize=(8, 5))
        sns.countplot(data=df, x=feature, palette='pastel')
        plt.title(f'Bar plot of {feature}')
        plt.xlabel(feature)
        plt.ylabel('Count')
        plt.xticks(rotation=45)
        plt.show()

        # Pie chart
        plt.figure(figsize=(8, 5))
        value_counts = df[feature].value_counts()
        value_counts.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=140)
        plt.title(f'Pie chart of {feature}')
        plt.ylabel('')
        plt.show()
```

```python
categorical_features = ['ADDRTYPE', 'COLLISIONTYPE', 'WEATHER', 'ROADCOND', 'LIGHTCOND']
```

```python
plot_categorical_features(df, categorical_features)    #function call
```

```python
def plot_bivariate_analysis(df):
    # Bivariate analysis with target column (severity description)
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x='SEVERITYDESC', hue='ADDRTYPE', palette='pastel')
    plt.title('Count of Accidents by Address Type and Severity Description')
    plt.xlabel('Severity Description')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.legend(title='Address Type')
    plt.show()
```

# 7 . Various Hypothesis Tests

a) The Chi-Square test is typically used for categorical data. Let's assume we are interested in testing if there is a significant relationship between ADDRTYPE and SEVERITYDESC.

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['ADDRTYPE'], df['SEVERITYDESC'])

# Perform the Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f'Chi-Square Statistic: {chi2}')
print(f'p-value: {p}')
```

b) The Shapiro-Wilk test tests the null hypothesis that the data was drawn from a normal distribution.

```python
from scipy.stats import shapiro

# Test for normality on a numerical feature
stat, p = shapiro(df['PERSONCOUNT'].dropna())
print('Statistics=%.3f, p=%.3f' % (stat, p))

# Interpret
alpha = 0.05
if p > alpha:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian (reject H0)')
```

c) The Kolmogorov-Smirnov test compares the sample distribution with a reference probability distribution ( such as normal distribution).

```python
from scipy.stats import kstest

# Test for normality on a numerical feature
stat, p = kstest(df['PERSONCOUNT'].dropna(), 'norm')
print('Statistics=%.3f, p=%.3f' % (stat, p))

# Interpret
if p > alpha:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian (reject H0)')
```

# 8 and 9. Correlation Testing and Summary

```python
from scipy.stats import pearsonr

# List of numerical features
numerical_features = ['PERSONCOUNT', 'PEDCOUNT', 'PEDCYLCOUNT', 'VEHCOUNT', 'INJURIES', 'SERIOUSINJURIES', 'FATALITIES']

# Calculate Pearson correlation for each pair of features
for i in range(len(numerical_features)):
    for j in range(i+1, len(numerical_features)):
        feature1 = numerical_features[i]
        feature2 = numerical_features[j]
        correlation_coefficient, p_value = pearsonr(df[feature1].dropna(), df[feature2].dropna())
        print(f"Pearson Correlation Coefficient between {feature1} and {feature2}: {correlation_coefficient:.4f}, p-value: {p_value:.4f}")

# Visualize relationships using scatterplots for relationships
for feature1 in numerical_features:
    for feature2 in numerical_features:
        if feature1 != feature2:
            plt.figure(figsize=(6, 4))
            sns.scatterplot(x=df[feature1], y=df[feature2])
            plt.title(f'Scatter plot between {feature1} and {feature2}')
            plt.xlabel(feature1)
            plt.ylabel(feature2)
            plt.show()
```

```python
# Create an empty correlation matrix DataFrame
correlation_matrix = pd.DataFrame(index=numerical_features, columns=numerical_features)

# Calculate Pearson correlation coefficient for each pair of numerical features
for feature1 in numerical_features:
    for feature2 in numerical_features:
        correlation_coefficient, _ = pearsonr(df[feature1], df[feature2])  # Using pearsonr function
        correlation_matrix.loc[feature1, feature2] = correlation_coefficient

# Convert correlation coefficients to numeric values
correlation_matrix = correlation_matrix.astype(float)

# Create a heatmap using Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='RdYlBu', fmt=".2f")  # Change the colormap to 'RdYlBu'
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

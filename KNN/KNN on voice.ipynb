```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('dataset/voice.csv')
```

```python
df.head()
```

```python
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
  

# df['label_encode']= label_encoder.fit_transform(df['label'])
  
# df['label_encode'].unique()
# Replace 0 with 1 and 1 with 0 in the specified column
df['label'] = df['label'].replace({'male': 0, 'female': 1})
```

```python
data=df.drop(columns=['label'])
```

```python
data.head()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
df_scaled = pd.DataFrame(data=scaled_data, columns=data.columns)
```

```python
df_scaled.describe()
```

```python
df_scaled.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_scaled,df['label'],test_size=0.2,random_state=43)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Instantiate the KNN model with default parameters
knn = KNeighborsClassifier()
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
df['label']
```

```python
params=knn.get_params()
print(params)
```

```python
knn.fit(x_train, y_train)
```

```python
preds=knn.predict(x_test.values)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, preds)
```

```python
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
# Compute the confusion matrix
cm = confusion_matrix(y_test, preds)

# Create a heatmap visualization of the confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d", cbar=False)
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import classification_report

# Assuming you have your true labels and predicted label

# Generate the classification report
report = classification_report(y_test, preds)
```

```python
print(report)
```
